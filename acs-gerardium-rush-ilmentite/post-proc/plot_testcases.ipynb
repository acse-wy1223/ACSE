```python
from main import read_data, create_diagram, annotate_image, display_image, create_table_image
import numpy as np
import pandas as pd

file_path = "../testcases/test_validity.csv"
df = pd.read_csv(file_path, header=None)

array_column = [list(map(int, x.split())) for x in df[0]]
bool_column = df[1].apply(lambda x: x.lower() == 'valid').tolist()
string_column = df[2].tolist()

def show_circuit(idx):

    circuit_conf = array_column[idx]
    circuit_conf = np.array(circuit_conf)
    performance_values = [0, 0, 0]
    performance_values = np.array(performance_values)
    
    if circuit_conf is not None and performance_values is not None:
        file = create_diagram(circuit_conf, show_labels=False)
        if file:
            annotated_file = annotate_image(file, circuit_conf,
                                            performance_values)
            if annotated_file:
                display_image(annotated_file)

for idx in range(len(array_column)):
    Label = "Valid" if bool_column[idx] else "Invalid"
    print(f"{Label}: {string_column[idx]}")
    show_circuit(idx)
```
